```
import pandas as pd
import statsmodels.formula.api as sms
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# Lê os dados
credit = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Credit.csv")
# Tira a primeira coluna
print(credit.columns)
credit = credit.drop(columns=['Unnamed: 0'])
```

```
# Após retirar coluna com índices...
print(credit.columns)
```

```
# Estimando a diferença entre o balance entre homens e mulheres usando regressão linear.
model1 = sms.ols(formula="Balance~Gender", data=credit).fit()
print(model1.summary())
```

```
# A variável Gender não é significativa para estudar a dívida média de cartões de crédito.
```

```
# Estimando a diferença entre o balance entre pessoas de etnias diferentes.
model2 = sms.ols(formula="Balance~Ethnicity", data=credit).fit()
print(model2.summary())
```

```
# A variável Ethnicity não é significativa para estudar a dívida média de cartões de crédito.
```

```
# Estimando um modelo utilizando todas as variáveis.
model3=sms.ols(formula="Balance~Income+Limit+Rating+Cards+Age+Education+Gender+Student+Married+Ethnicity", data=credit).fit()
print(model3.summary())
```

```
# Muitas variáveis são insignificantes para explicar o modelo.
# Não. Apenas três.
# Eu tiraria as variáveis: Gender, Married, Ethnicity, Limit, Cards, Age e Education.
# Ficaria com as variáveis: Income, Rating e Student.
```

```
# Estimando modelo melhorado
model4=sms.ols(formula="Balance~Income+Rating+Student", data=credit).fit()
print(model4.summary())
```

```
# Todas as variáveis do novo modelo são significativas e o modelo tem um grau de explicação bem satisfatório.
```
