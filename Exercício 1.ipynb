```
import pandas as pd
import statsmodels.formula.api as sms
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# Lendo os dados
boston = pd.DataFrame.from_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
boston.head()
```

```
print(boston.columns)
```

```
# Estimando a primeira regressão
model1 = sms.ols(formula="medv~lstat", data=boston).fit()
print(model1.summary())
```

```
# Estimando a segunda regressão
model2 = sms.ols(formula="medv~age", data=boston).fit()
print(model2.summary())
```

```
# Estimando a terceira regressão
model3 = sms.ols(formula="medv~lstat+age", data=boston).fit()
print(model3.summary())
```

```
# No primeiro modelo, a variável lstat é significativa e bastante explicativa para prever medv.
# No segundo modelo, a variável age é significativa, porém, explica muito pouco o medv.
# No terceiro modelo, as duas variáveis são significativas, porém, comparado com a variável do primeiro modelo, a variável age não ajuda a explicar medv.
```

```
# Estimando regressão com todas as variáveis
model4 = sms.ols(formula="medv~zn+indus+chas+nox+rm+age+dis+rad+tax+ptratio+b+lstat", data=boston).fit()
print(model4.summary())
```

```
# Não. Eu tiraria as variáveis indus e age pois elas não são significativas para prever medv. O ajuste foi considerável, com um grau de explicação de 73%.
```

```
# Estimando regressão tirando as variáveis indus e age
model5 = sms.ols(formula="medv~zn+chas+nox+rm+dis+rad+tax+ptratio+b+lstat", data=boston).fit()
print(model5.summary())
```

```
# O grau de explicação continuou praticamente o mesmo, porém, agora todas as variáveis são significativas para prever medv.
```
